### get intrinsics K and extrinsics E = R|t

```python
#find the matrix K
def get_intrinsic_parameters(H_r):
        M = len(H_r)
        V = np.zeros((2*M, 6), np.float64)

        def v_pq(p, q, H):
            v = np.array([
                    H[0, p]*H[0, q],
                    H[0, p]*H[1, q] + H[1, p]*H[0, q],
                    H[1, p]*H[1, q],
                    H[2, p]*H[0, q] + H[0, p]*H[2, q],
                    H[2, p]*H[1, q] + H[1, p]*H[2, q],
                    H[2, p]*H[2, q]
                ])
            return v

        for i in range(M):
            H = H_r[i]
            V[2*i] = v_pq(p=0, q=1, H=H)
            V[2*i + 1] = np.subtract(v_pq(p=0, q=0, H=H), v_pq(p=1, q=1, H=H))

        # solve V.b = 0
        u, s, vh = np.linalg.svd(V)
        b = vh[np.argmin(s)]       
        print("b", b)
        B = [b[0],b[1],b[3],b[1],b[2],b[4],b[3],b[4],b[5]]
        B = np.array(B).reshape(3,3)
        print("B",B)
        
        vc = (b[1]*b[3]-b[0]*b[4])/(b[0]*b[2]-b[1]**2)
        l= b[5]-(b[3]**2+vc*(b[1]*b[2]-b[0]*b[4]))/b[0]
        alpha = np.sqrt((l/b[0]))
        beta = np.sqrt(((l*b[0])/(b[0]*b[2]-b[1]**2)))
        gamma=-1*((b[1])*(alpha**2)*(beta/l))
        uc=(gamma*vc/beta)-(b[3]*(alpha**2)/l)
        K = np.array([
        [alpha, gamma, uc],
        [0,     beta,  vc],
        [0,     0,      1]
        ])
        return K
```

```python
#find the matrix E 
def estimate_view_transform(intrinsics, homography):

    #homography = homography.reshape(3, 3)
    inv_intrinsics = np.linalg.inv(intrinsics)

    h1 = homography[:, 0]
    h2 = homography[:, 1]
    h3 = homography[:, 2]

    ld1 = 1 / np.linalg.norm(np.dot(inv_intrinsics, h1))
    ld2 = 1 / np.linalg.norm(np.dot(inv_intrinsics, h2))
    ld3 = (ld1 + ld2) / 2

    r0 = ld1 * np.dot(inv_intrinsics, h1)
    r1 = ld2 * np.dot(inv_intrinsics, h2)
    r2 = np.cross(r0, r1)

    t = np.array(ld3 * np.dot(inv_intrinsics, h3)).transpose()
    
    Rt = np.array(
            [r0.transpose(), r1.transpose(), r2.transpose(), t.transpose()]
         ).transpose()
    
    #R_ort = np.array([r0.transpose(), r1.transpose(), r2.transpose()])
    #u, s, v = np.linalg.svd(R_ort)
    #R_ort = np.dot(u,v.transpose())
    
    # todo: Add optimization?
    #Rt = np.array(R_ort[0],R_ort[1],R_ort[2],t.transpose()).transpose()
    
    return Rt


def get_camera_extrinsics(intrinsics, homographies):
    extrinsics = []
    for i in range(0, len(homographies)):
        extrinsics.append(estimate_view_transform(intrinsics, homographies[i]))

    return extrinsics
```

```python
from scipy.optimize import fsolve
from scipy.optimize import newton_krylov
import math

#estimates k1 and k2 
def estimate_lens_distortion(intrinsics, P, corners, world):
    #goal create A and b
    #K intrinsics contains (u_0,v_0,a_u,a_v)
    u_0 = intrinsics[0][2]
    v_0 = intrinsics[1][2]
    a_u = intrinsics[0][0]
    a_v = intrinsics[1][1]
    #calculate r_2 = ((u-u_0)/a_u)^2 + ((u-u_0)/a_u)^2
    # lets find u and v and crate A and b
    b = np.array([])
    A = []
    radiuses = np.array([])
    for i in range(0 , len(world)):
        m_i = world[i]
        m_i = (m_i[0],m_i[1],0,1)
        #print("mi_i ",m_i)
        u_i = corners[i][0][0]
        v_i = corners[i][0][1]
        # u = p1'm_i/p3'm_i
        u = np.dot(P[0],m_i)/np.dot(P[2],m_i)
        #print("predicted u ",u,  "Real u ", u_i )
        #(p2'm_i/p3'm_i - v_i)^2
        v = np.dot(P[1],m_i)/np.dot(P[2],m_i)
        #print("predicted v ", v, "Real v ", v_i)
        r_2 = math.pow(((u-u_0)/a_u),2) + math.pow(((v-v_0)/a_v),2)
        r_4 = r_2 * r_2
        radiuses = np.append(radiuses, r_2)
        u_hat = u_i # is the u corner detected
        v_hat = v_i # is the v corner detected
        #create b
        b = np.append(b, u_hat-u)
        b = np.append(b, v_hat-v)
        #crate A
        A.append([((u-u_0)*r_2), ((u-u_0)*r_4)])
        A.append([((v-v_0)*r_2), ((v-v_0)*r_4)])
        
    A = np.array(A)
    #print("A", A)
    #print("b", b[7])
    #A'A)..A'b
    ATAI = np.dot(A.transpose(),A)
    ATAI = np.linalg.inv(ATAI)
    
    ATAI = np.dot(ATAI,A.transpose())
    
    ATAI = np.dot(ATAI,b)
    #print("ATAI",ATAI)
    k = np.linalg.lstsq(A, b)
        
    return (k,radiuses)



# solve non linear equation to find new_u and new_v given (k1,k2 and corners)
def getUandV(intrinsics, k1, k2, corners,radiuses,initial_corner_set):  
    def equations(p):
        u, v = p
        return ((u-u_0)*(r_2)*k1 + (u-u_0)*(r_4)*k2+u-u_hat  ,(v-v_0)*(r_2)*k1 + (v-v_0)*(r_4)*k2 + v-v_hat)
    
    u_0 = intrinsics[0][2]
    v_0 = intrinsics[1][2]
    a_u = intrinsics[0][0]
    a_v = intrinsics[1][1]
    #for every point 
    print ("len(corners)",len(initial_corner_set) )
    new_pixels = []
    for i in range(0 , len(initial_corner_set)):
        u_hat = initial_corner_set[i][0][0] # is the u corner detected
        v_hat = initial_corner_set[i][0][1] # is the v corner detected
        r_2 = radiuses[i]
        r_4 = r_2 * r_2
        try:  
            u_new, v_new =  newton_krylov(equations, (-1, -1))   
        except:  
            #print ("First try FAILED" )
            try: 
                u_new, v_new =  newton_krylov(equations, (0, 0))

            except: 
                #print("SECOND try FAILED" )
                try: 
                    u_new, v_new =  newton_krylov(equations, (1, 1))
                except:  
                    print("LAST try FAILED")
                    u_new = None
                    v_new = None
        new_pixels.append([[u_new,v_new]])
    new_pixels = np.array(new_pixels)
    return new_pixels    


def GetNewPixels(world_set ,corners_set,initial_corner_set):
    #calcola H per ogni set di corners (relativo ad un immagine)
    homos = []
    newPixels = []
    PS = []
    for i in range(0,len(world_set)):
        H = getHomographie(corners_set[i],world_set[i])
        homos.append(H)

    #voglio K per ogni gruppo
    K = get_intrinsic_parameters(homos)

    #voglio R E T 
    #il calcolo va fatto per n immagini e vanno trovate  n P 
    for i in range(0,len(world_set)):
        E = estimate_view_transform(K, homos[i])
        #voglio P
        P = getP(E,K)
        PS.append(P)
        #calcola i coeficienti k1 e k2 per ogni immagine
        D = estimate_lens_distortion(K, P, corners_set[i],world_set[i])
        k1 = D[0][0][0]
        k2 = D[0][0][1]
        radiuses = D[-1]
        # solve non linear equation to find new_u and new_v given (k1,k2 and corners)
        new_pixel = getUandV(K, k1, k2, corners_set[i], radiuses,initial_corner_set[i]) 
        newPixels.append(new_pixel)
        #print(newPixels)
    return (newPixels,k1,k2,K,PS)


```
